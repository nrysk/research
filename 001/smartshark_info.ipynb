# SmartSHARK Information

SmartSHARK データの基本情報を調査する

### Commit

| Column | Description |
| --- | --- |
| project | name of the project |
| nc |  # of commits |
| fcd | first commit date |
| lcd | last commit date |
| nbfc |  # of bug-fixing commits |
| nbfc_a |  # of bug-fixing commits with adjustedszz |
| nbfc_io |  # of bug-fixing commits with issueonly |
| nbfc_v |  # of bug-fixing commits with validated |
| nbfc_if |  # of bug-fixing commits with issuefasttext |
| nbic |  # of bug-inducing commits |
| nbic_szz |  # of bug-inducing commits with SZZ |
| nbic_jl+r |  # of bug-inducing commits with JL+R |
| nbic_jlip+r |  # of bug-inducing commits with JLIP+R |
| nbic_jlmiv |  # of bug-inducing commits with JLMIV |
| nbic_jlmiv+ |  # of bug-inducing commits with JLMIV+ |
| nbic_jlmiv+av |  # of bug-inducing commits with JLMIV+AV |
| nbic_jlmiv+rav |  # of bug-inducing commits with JLMIV+RAV |
| nbic_jlmiv+r |  # of bug-inducing commits with JLMIV+R |


```python
import polars as pl

pl.Config(tbl_rows=100)

df = pl.read_csv("data/commit_info.csv")
# 合計行を追加
row = {col: df[col].sum() for col in df.columns if df[col].dtype == pl.Int64}
row["project"] = "total"
df = pl.concat([df, pl.DataFrame(row)], how="diagonal")
display(df)
```

### Pull Request
