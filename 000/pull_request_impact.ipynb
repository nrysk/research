# Pull Request Effect

バグ混入にプルリクエストが影響するかを調査する．

```python
import polars as pl
import polars.selectors as sel

pl.Config(tbl_rows=100)

(
    pl.read_csv("data/pull_request_impact.csv")
    .select("project", "#pr", "#mpr", sel.ends_with("all"), sel.ends_with("jl+r"))
    .sort("#cmt-pr+bi_jl+r")
)
```

## PR と bug の数

```python
import matplotlib.pyplot as plt
import polars as pl


def print_result(df: pl.DataFrame) -> None:
    sum_dict = {
        "#cmt+pr": df["#cmt+pr+bi_all"].sum() + df["#cmt+pr-bi_all"].sum(),
        "#cmt-pr": df["#cmt-pr+bi_all"].sum() + df["#cmt-pr-bi_all"].sum(),
    }
    print(f"#cmt+pr: {sum_dict['#cmt+pr']}")
    print(f"#cmt-pr: {sum_dict['#cmt-pr']}")
    for label in ["all", "szz", "jl+r"]:
        print(f"--- {label} ---")
        print(
            f"#cmt+pr+bi: {df[f"#cmt+pr+bi_{label}"].sum()} ({df[f"#cmt+pr+bi_{label}"].sum() / sum_dict["#cmt+pr"] * 100:.1f}% of #cmt+pr)"
        )
        print(
            f"#cmt+pr-bi: {df[f"#cmt+pr-bi_{label}"].sum()} ({df[f"#cmt+pr-bi_{label}"].sum() / sum_dict['#cmt+pr'] * 100:.1f}% of #cmt+pr)"
        )
        print(
            f"#cmt-pr+bi: {df[f"#cmt-pr+bi_{label}"].sum()} ({df[f"#cmt-pr+bi_{label}"].sum() / sum_dict['#cmt-pr'] * 100:.1f}% of #cmt-pr)"
        )
        print(
            f"#cmt-pr-bi: {df[f"#cmt-pr-bi_{label}"].sum()} ({df[f"#cmt-pr-bi_{label}"].sum() / sum_dict['#cmt-pr'] * 100:.1f}% of #cmt-pr)"
        )


df = pl.read_csv("data/pull_request_impact.csv")

# プロジェクト全体のバグの数と割合を算出
print(f"# of projects: {len(df)}")
print_result(df)
print("-" * 80)

# プルリクエストを経由したコミットの数が 0 のプロジェクトを除外
filtered_df = df.filter(pl.col("#cmt+pr+bi_all") + pl.col("#cmt+pr-bi_all") > 0)
print(f"# of projects: {len(filtered_df)}")
print_result(filtered_df)
print("-" * 80)

# プルリクエストを経由したコミットの数が 100 以下のプロジェクトを除外
filtered_df = df.filter(pl.col("#cmt+pr+bi_all") + pl.col("#cmt+pr-bi_all") > 100)
print(f"# of projects: {len(filtered_df)}")
print_result(filtered_df)
print("-" * 80)

# jl+r によるバグの数が 0 のプロジェクトを除外
filtered_df = df.filter(pl.col("#cmt-pr+bi_jl+r") > 0)
print(f"# of projects: {len(filtered_df)}")
print_result(filtered_df)
print("-" * 80)
```

## PR と bug の箱ひげ図

```python
import matplotlib.pyplot as plt
import polars as pl
import seaborn as sns


def plot_bug_rate(df: pl.DataFrame):
    """バグ率の箱ひげ図をプロットする関数"""
    # 各プロジェクトに対して, バグの割合の列を追加
    szz_labels = ["all", "szz", "jl+r"]
    for label in szz_labels:
        df = df.with_columns(
            [
                (
                    pl.col(f"#cmt+pr+bi_{label}")
                    / (pl.col(f"#cmt+pr+bi_{label}") + pl.col(f"#cmt+pr-bi_{label}"))
                ).alias(f"%bug+pr_{label}"),
                (
                    pl.col(f"#cmt-pr+bi_{label}")
                    / (pl.col(f"#cmt-pr+bi_{label}") + pl.col(f"#cmt-pr-bi_{label}"))
                ).alias(f"%bug-pr_{label}"),
            ]
        )

    # プロジェクトごとのバグ割合を箱ひげ図で表示
    plt.rc("font", family="Noto Sans CJK JP")
    plt.figure(figsize=(10, 5))
    x = [df[f"%bug+pr_{label}"] for label in szz_labels] + [
        df[f"%bug-pr_{label}"] for label in szz_labels
    ]
    tick_labels = [f"%bug+pr_{label}" for label in szz_labels] + [
        f"%bug-pr_{label}" for label in szz_labels
    ]
    plt.boxplot(x, tick_labels=tick_labels)
    plt.ylim(0 - 0.1, 1 + 0.1)
    plt.show()


df = pl.read_csv("data/pull_request_impact.csv")

# 各プロジェクトに対して, バグ率の箱ひげ図をプロット
print(f"# of projects: {len(df)}")
plot_bug_rate(df)

# プルリクエストを経由したコミット数が 0 のプロジェクトを除外
filtered_df = df.filter(pl.col("#cmt+pr+bi_all") + pl.col("#cmt+pr-bi_all") > 0)
print(f"# of projects: {len(filtered_df)}")
plot_bug_rate(filtered_df)

# プルリクエストを経由したコミット数が 100 以下のプロジェクトを除外
filtered_df = df.filter(pl.col("#cmt+pr+bi_all") + pl.col("#cmt+pr-bi_all") > 100)
print(f"# of projects: {len(filtered_df)}")
plot_bug_rate(filtered_df)

# jl+r によるバグの数が 0 のプロジェクトを除外
filtered_df = df.filter(pl.col("#cmt-pr+bi_jl+r") > 0).filter(
    pl.col("#cmt+pr+bi_all") + pl.col("#cmt+pr-bi_all") > 0
)
print(f"# of projects: {len(filtered_df)}")
plot_bug_rate(filtered_df)
```

```python
"""
jl+r によってバグ混入のラベル付けを行ったプロジェクトを対象に
バグ混入コミットの割合を示す箱ひげ図をプロット
"""

import matplotlib.pyplot as plt
import polars as pl

# jl+r によるバグの数が 0 のプロジェクトを除外
df = pl.read_csv("data/pull_request_impact.csv").filter(pl.col("#cmt-pr+bi_jl+r") > 0)
print(f"# of projects: {len(df)}")

# プルリクエスト有りのバグコミット割合を算出
bug_rates_in_pr = df.filter(
    pl.col("#cmt+pr+bi_jl+r") + pl.col("#cmt+pr-bi_jl+r") > 0
).with_columns(
    (
        pl.col("#cmt+pr+bi_jl+r")
        / (pl.col("#cmt+pr+bi_jl+r") + pl.col("#cmt+pr-bi_jl+r"))
        * 100
    ).alias("bug_rate")
)[
    "bug_rate"
]

# プルリクエスト無しのバグコミット割合を算出
bug_rates_not_in_pr = df.with_columns(
    (
        pl.col("#cmt-pr+bi_jl+r")
        / (pl.col("#cmt-pr+bi_jl+r") + pl.col("#cmt-pr-bi_jl+r"))
        * 100
    ).alias("bug_rate")
)["bug_rate"]

# プロジェクトごとのバグ割合を箱ひげ図で表示
plt.figure(figsize=(4, 3))
plt.boxplot(
    [bug_rates_in_pr, bug_rates_not_in_pr],
    tick_labels=["in PR", "not in PR"],
    widths=0.3,
)
plt.ylabel("% of bug commits")
plt.ylim(0 - 1, 100 + 1)
plt.show()
```

### PR と bug の検定

```python
import math

import numpy as np
import polars as pl
from scipy.stats import chi2_contingency, fisher_exact

# ファイル読み込み
df = pl.read_csv("data/pull_request_impact.csv").filter(pl.col("#cmt-pr+bi_jl+r") > 0)
print(f"# of projects: {len(df)}")
# テーブルの作成
table = np.array(
    [
        [
            df["#cmt+pr+bi_jl+r"].sum(),
            df["#cmt+pr-bi_jl+r"].sum(),
        ],
        [
            df["#cmt-pr+bi_jl+r"].sum(),
            df["#cmt-pr-bi_jl+r"].sum(),
        ],
    ]
)

# カイ二乗検定
chi2, p, dof, expected = chi2_contingency(table)
phi = math.sqrt(chi2 / sum(table.flatten()))

# 結果の表示
print("--- Chi-squared test ---")
print("table:")
print(table)
print(f"chi2: {chi2}")
print(f"p-value: {p}")
print(f"degree of freedom: {dof}")
print(f"phi: {phi}")
print("expected:")
print(expected)

# フィッシャーの正確確率検定
oddsratio, p = fisher_exact(table)

# 結果の表示
print("--- Fisher's exact test ---")
print(f"odds ratio: {oddsratio}")
print(f"p-value: {p}")
```
