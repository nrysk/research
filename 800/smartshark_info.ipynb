```python
import polars as pl

pl.Config(tbl_rows=100)

df = pl.read_csv("data/commit_info.csv")
# 合計行を追加
row = {col: df[col].sum() for col in df.columns if df[col].dtype == pl.Int64}
row["project"] = "total"
df = pl.concat([df, pl.DataFrame(row)], how="diagonal")
display(df)

# 修正コミットが存在しないプロジェクトを抽出
df = df.filter(pl.col("nbfc") == 0)
for project in df["project"]:
    print(project)
```

### データの整合性の確認

```python
"""
pull_request_commit と commit が両方存在することを確認
"""

import os

from mongoengine import connect
from pycoshark.mongomodels import (
    Commit,
    FileAction,
    Project,
    PullRequestCommit,
    VCSSystem,
)
from pycoshark.utils import create_mongodb_uri_string

# データベースに接続
uri = create_mongodb_uri_string(
    db_user=os.getenv("SMARTSHARK_DB_USERNAME"),
    db_password=os.getenv("SMARTSHARK_DB_PASSWORD"),
    db_hostname=os.getenv("SMARTSHARK_DB_HOST"),
    db_port=os.getenv("SMARTSHARK_DB_PORT"),
    db_authentication_database=os.getenv("SMARTSHARK_DB_AUTHENTICATION_DATABASE"),
    db_ssl_enabled=False,
)
connect(
    db=os.getenv("SMARTSHARK_DB_DATABASE"),
    host=uri,
)

pull_request_commits = PullRequestCommit.objects
pull_request_commit_count = pull_request_commits.count()

for i, pull_request_commit in enumerate(pull_request_commits):
    # 進捗表示
    if i % 1000 == 0:
        print(f"{i}/{pull_request_commit_count}", end="\r")

    commit_by_hash = Commit.objects(
        revision_hash=pull_request_commit.commit_sha
    ).first()

    if pull_request_commit.commit_id is None and commit_by_hash is not None:

        print(
            f"pull_request_commit_id: {pull_request_commit.id}, commit_id: {pull_request_commit.commit_id}, commit_sha: {pull_request_commit.commit_sha}"
        )
```
